```python
import pandas as pd

sc_with_mttu_mttr = pd.read_csv("../step4_mttu_mttr_and_sc_data/sc_mttu_mttr_final.csv")
sc_wo_mttu_mttr = pd.read_csv("../../data/local_nonlocal_checks_with_missing_values.csv")
control_var_data = pd.read_csv("../step1_control_var/control_data_final.csv")
```

## Merging Control Variable with SC Data (No MTTU / MTTR)

Note that this dataset does not merge on the 'cleaned' tag name.

```python
sc_wo_mttu_mttr = sc_wo_mttu_mttr.drop(columns=['package_name', 'project_name', 'published_at', 'package_name', 'githubrepo'])
sc_control_no_mttu_mttr = pd.merge(sc_wo_mttu_mttr, 
                       control_var_data,
                       left_on=['github_repo', 'tag_name', 'tag_commit_sha'], 
                       right_on=['github_repo', 'tag_name', 'tag_commit_sha'], 
                       how='inner')
```

```python
sc_control_no_mttu_mttr = sc_control_no_mttu_mttr.drop_duplicates(subset=['github_repo', 'tag_name', 'tag_commit_sha'])
```

```python
sc_control_no_mttu_mttr.to_csv("sc_control_data_final.csv", index=False)
```

## Merging Control Variable with SC and MTTU / MTTR Data 

Note that this dataset does not merge on the 'cleaned' tag name.

```python
sc_with_mttu_mttr = sc_with_mttu_mttr.drop(columns=['package_name', 'project_name', 'published_at', 'package_name', 'githubrepo', 'window_start', 'window_end', 'status'])
sc_control_with_mttu_mttr = pd.merge(sc_with_mttu_mttr, 
                       control_var_data,
                       left_on=['github_repo', 'tag_name', 'tag_commit_sha'], 
                       right_on=['github_repo', 'tag_name', 'tag_commit_sha'], 
                       how='inner')

sc_control_with_mttu_mttr_no_duplicates = sc_control_with_mttu_mttr.drop_duplicates(subset=['github_repo', 'tag_name', 'tag_commit_sha'])
```

```python
sc_control_with_mttu_mttr_no_duplicates.to_csv("sc_control_mttu_mttr_data_final.csv", index=False)
```
