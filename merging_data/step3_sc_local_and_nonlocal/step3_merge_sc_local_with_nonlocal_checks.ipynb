Combine the maintained, contributors, and dependency update tools with the other checks. 

```python
import pandas as pd

non_local_data = pd.read_csv("step2_merge_local_and_nonlocal/final_nonlocal_data.csv")
local_data = pd.read_csv("scorecard_local_metrics_final.csv")
```

```python
## remove some unnecessary columns from local data
local_data = local_data.drop(columns=['source_file', 'aggregate_score', 'run_timestamp'])
```

```python
## remove any rows with github_repo values that are not in the non_local data
local_data1 = local_data[local_data['github_repo'].isin(non_local_data['githubrepo'])]
```

```python
## need to merge local and non_local data on github_repo, tag_name, and tag_commit_sha
non_local_data =non_local_data.drop(columns=["published_at", "status", "release_tag_name", "project_name"])

merged_data_1 = pd.merge(local_data1,
                         non_local_data,
                         left_on=['github_repo', 'tag_name', 'tag_commit_sha'],
                         right_on=['githubrepo', 'tag_name', 'tag_commit_sha'],
                         how = 'outer')

merged_data_2 = pd.merge(local_data1,
                         non_local_data,
                         left_on=['github_repo', 'tag_name', 'tag_commit_sha'],
                         right_on=['githubrepo', 'tag_name', 'tag_commit_sha'],
                         how = 'inner')

## remove duplicates
final_merged_data = merged_data_2.drop_duplicates(subset=['github_repo', 'tag_name', 'tag_commit_sha'], keep='first')
```

```python
## save the dataset that includes the missing values 
final_merged_data.to_csv("local_nonlocal_checks_with_missing_values.csv", index=False)
```
